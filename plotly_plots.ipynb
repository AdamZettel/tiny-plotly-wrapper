```python
# Importing necessary libraries
import numpy as np
from plotly_wrapper import PlotlyWrapper
# Using the Plotly Wrapper
plt = PlotlyWrapper()
```

```python
# Line Plot Data
x_line = np.linspace(0, 10, 100)
y_line = np.sin(x_line)

# Basic Line Plot
plt.plot(x_line, y_line, label='Sine Wave')
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.title('Sine Wave')
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)
plt.legend()
plt.set_size(10, 6)
plt.show()
```

```python
# Scatter Plot Data
x_scatter = np.random.rand(100)
y_scatter = np.random.rand(100)
color_scatter = np.random.choice(['A', 'B', 'C', 'D'], size=100)

# Basic Scatter Plot
plt.scatter(x_scatter, y_scatter, label='Random Scatter', color='blue')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Scatter Plot')
plt.legend()
plt.set_size(10, 6)
plt.show()
```

```python
# Bar Plot Data
categories = ['Apples', 'Bananas', 'Cherries', 'Dates', 'Elderberries']
values = [10, 15, 7, 5, 12]

# Basic Bar Plot
plt.bar(categories, values, label='Fruit Count', color='orange')
plt.xlabel('Fruit')
plt.ylabel('Count')
plt.title('Fruit Count')
plt.legend()
plt.set_size(10, 6)
plt.show()
```

```python
# Histogram Data
data_hist = np.random.normal(size=1000)

# Basic Histogram
plt.hist(data_hist, bins=30, label='Random Normal Distribution', color='green')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Random Normal Distribution')
plt.legend()
plt.set_size(10, 6)
plt.show()
```

```python
# 3D Scatter Plot Data
x_3d = np.random.rand(100)
y_3d = np.random.rand(100)
z_3d = np.random.rand(100)

# Basic 3D Scatter Plot
plt.plot_3d(x_3d, y_3d, z_3d, label='3D Scatter', color='purple')
plt.title('3D Random Scatter Plot')
plt.set_size(10, 6)
plt.show()
```
